```python
from sklearn.cluster import KMeans
import pandas 
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("Mall_Customers.csv",engine='python',encoding='ISO-8859-1')

data.replace({'Male':0, 'Female':1}, inplace=True)
data = data.drop(data.columns[0], axis=1)

model = KMeans(n_clusters= 4, max_iter=1000)
model.fit(data)

```

```python
data_ = data[['Gender' , 'Age','Annual Income (k$)','Spending Score (1-100)' ]]
data_.head(20).T
```

```python
intertia = {}
for k in range(1,30):
    kmeans = KMeans(n_clusters=k,max_iter=1000)
    kmeans.fit(data)
    intertia[k] = kmeans.inertia_
plt.figure()
plt.plot(list(intertia.keys()),list(intertia.values()))
plt.xlabel("Number of Clusters")
plt.ylabel("inertia")
plt.show()
```
